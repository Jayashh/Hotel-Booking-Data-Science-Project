#Data Analysis and Visualization

```python
#Calculating the overall cancelled percentage accross all hotels
cancelled_percentage = df["is_canceled"].value_counts(normalize = True)
print(cancelled_percentage)

plt.figure(figsize = (5, 4))
plt.title("Cancelled vs Non-Cancelled Reservations")
# Creating a bar plot to check the count of cancellations accross both hotels
plt.bar(["Non cancelled reservations", "Cancelled reservations"], df["is_canceled"].value_counts(), edgecolor = "k", width = 0.7)

plt.ylabel("Number of reservations")
plt.show()
```

```python
#Creating a countplot of nominal/categorical df["hotel"] values in the "x" axis, and the "y" values which is the numerical count of reservations
plt.figure(figsize=(5, 4))
ax1 = sns.countplot(x = "hotel", hue = "is_canceled" , data = df, palette = "Blues")


legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1, 1))

plt.title("Number of cancelled reservations in Resort vs City Hotel")
plt.xlabel("Hotel type")
plt.ylabel("Reservation Count")
plt.legend(["non-cancelled", "cancelled"])
plt.show()
```

```python
#Creating a separate df to distinguish all the "Resort Hotels" from the original df
resort_hotel = df[df["hotel"] == "Resort Hotel"]

#Finding the ratio of non cancelled reservations to cancelled reservations for resort hotels
resort_hotel["is_canceled"].value_counts(normalize = True)
```

```python
#Creating a separate df to distinguish all the "Resort Hotels" from the original df
city_hotel = df[df["hotel"] == "City Hotel"]

#Finding the ratio of non cancelled reservations to cancelled reservations for city hotels
city_hotel["is_canceled"].value_counts(normalize = True)
```

```python
#Grouping by the unique reservation date (categorical ordinal variable) and adr (continious aggregated variable)
resort_hotel = resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel = city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize = (10, 4))
plt.title("Average daily rate in City vs Resort Hotels", fontsize = 30)

#Creating a line plot to compare the ADR for resort vs city hotels.
plt.plot(resort_hotel.index, resort_hotel["adr"], label = "Resort Hotel")
plt.plot(city_hotel.index, city_hotel['adr'], label = "City Hotel")


plt.xlabel("Different days")
plt.ylabel("Average Daily Rate")
plt.legend(fontsize = 20)
plt.show()
```

```python
#Create a countplot to compare months (discreet categorical variable) with count of is_cancelled
df["month"] = df["reservation_status_date"].dt.month

plt.figure(figsize = (8, 4))
ax1 = sns.countplot(x = "month", hue = "is_canceled", data = df, palette = "bright")

legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1, 1))
plt.title("Revervation status (count of is_cancelled) per month", size = 20)
plt.xlabel("Months")
plt.ylabel("Reservation status")
plt.legend(["Not cancelled", "Cancelled"])
plt.show()
```

```python
#Creating a bar plot in which we compare every unique month (x axis) to the total ADR (y axis) for only the cancelled reservations
plt.figure(figsize = (8, 4))
plt.title("Total ADR for Months for Cancelled Reservations", fontsize = 20)

sns.barplot(x = "month", y = "adr", data = df[df["is_canceled"] == 1].groupby("month")[["adr"]].sum().reset_index())
plt.show()
```

```python
#Top 10 countries with the highest cancellation rate
cancelled_data = df[df["is_canceled"] == 1]
top_10_countries = cancelled_data["country"].value_counts()[:10]
top_10_countries
```

```python
#Create a pie-chart which represents the ratio of cancellation rate accross top 10 countries
plt.figure(figsize = (5, 5))
plt.title("Cancellation rate accross top 10 countries", fontsize = 20)

plt.pie(top_10_countries, autopct = "%.2f", labels = top_10_countries.index)
plt.show()
```

```python
#Checking if client comes from offline or online method (for all bookings)
df["market_segment"].value_counts()
```

```python
#Checking the method of booking for just the cancelled bookings

cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
#Grouping by unique reservation dates and average adr for that period
cancelled_df_adr = cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
cancelled_df_adr.reset_index(inplace = True)

#Sorting reservation status date by ascending order
cancelled_df_adr.sort_values("reservation_status_date", inplace = True)
cancelled_df_adr
```

```python
non_cancelled_data = df[df["is_canceled"] == 0]

non_cancelled_df_adr = non_cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
non_cancelled_df_adr.reset_index(inplace = True)

non_cancelled_df_adr.sort_values("reservation_status_date", inplace = True)
non_cancelled_df_adr
```

If we reset the index using inplace = True, we can access the column name using df["column"], but if the index is not reset, then we need to access the column using df.index


```python
#Plotting the adr for cancelled reservations vs non-cancelled reservations
```

```python
plt.figure(figsize = (10, 4))
plt.title("adr for cancelled reservations vs non-cancelled reservations",  fontsize = 20)

plt.plot(non_cancelled_df_adr["reservation_status_date"], non_cancelled_df_adr["adr"], label = "Non-Cancelled")
plt.plot(cancelled_df_adr["reservation_status_date"], cancelled_df_adr["adr"], label = "Cancelled")
plt.legend()
plt.show()
```

```python
#Filter the graph above to just show data from 2016 to September 2017. We will do this by updating the reservation_status_date column
non_cancelled_df_adr = non_cancelled_df_adr[(non_cancelled_df_adr["reservation_status_date"] >= "2016") & (non_cancelled_df_adr["reservation_status_date"] <= "2017-09-01") ]
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr["reservation_status_date"] >= "2016") & (cancelled_df_adr["reservation_status_date"] <= "2017-09-01")]
```

```python
#Plotting the new graph relating to the new desired dates

plt.figure(figsize=(20, 5))

plt.plot(non_cancelled_df_adr["reservation_status_date"], non_cancelled_df_adr["adr"], label= "Non-Cancelled ADR")
plt.plot(cancelled_df_adr["reservation_status_date"], cancelled_df_adr["adr"], label = "Cancelled ADR")

plt.title("Updated adr graph for cancelled vs non cancelled reservations", fontsize = 20)

plt.legend()
plt.show()
```
